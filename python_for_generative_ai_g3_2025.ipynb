```python
def pass_student(**kwargs):
    pass_student_list = []
    for name, marks in kwargs.items():
        if marks > 35:
            pass_student_list.append(name)
    return pass_student_list


```

```python
# supervised machine learning , it is a type of machine learning that deal with label data , it is solved two types of problem
    # 1)Regression based problem
    # 2)classification based problem
# label data-- there are two types of featured
#      input feature or independent feature             output feature or dependent feature or target value
#       ex- package prediction based on cgpa
            # cgpa  package
            # 10      9.5
            # 7       6.5
            # 8       7.2

      # Regression based problem -- in regression based problem the data type of output feature is continuous
      #         1) package prediction
      #                                          cgpa  package
      #                                           10      9.5
      #                                            9       8.4    (continuous)
      #                                            8       7.2


      #         2) house price prediction
      #                                           i/p                    o/p
      #                                           no. of room           price
      #                                           8                        15
      #                                           6                        12




```

```python
# classification based problem -- in this problem the o/p feature data type is discrete form
# ex -- result prediction               ex -- spam mail prediction
#           marks     result
#           70        pass- 1
#           32        fail - 0
#           65        pass - 1
```

```python
# unsupervised machine learning -- it is a type of machine learning in which deal with unlabel data .
# unsupervised machine learning solved following types of problem
# 1) clustering - grouping
# 2) dimensional reduction
# 3) anomaly detection
# 4) association
```

```python
# supervised machine learning algorithm solved regression based problem
# 1) linear regression
# 2)ridge regression
# 3) lasso regression
# 4) support vector regression
# 5) decision tree regression
# 6) random forest regression
```

```python
# supervised machine learning algorithm solved classification based problem
# 1) logistic regression
# 2) support vector classification
# 3) decision tree classification
# 4) random forest classification
```

```python
import numpy as np
from numpy.linalg import norm

A = np.array([2, 1, 2, 3, 2, 9])
B = np.array([3, 4, 2, 4, 5, 5])

# compute cosine similarity
cosine = np.dot(A, B) / (norm(A) * norm(B))
print("Cosine Similarity: ",cosine)
```
